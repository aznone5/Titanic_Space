# Dealing with null values
the submission requires the same amount of rows as the test set(4277) so we can't take out any rows that have null values.  The best thing to do is to give them an a numerical average, or a random string/boolean value that would work.

```python
def handle_null_values(df):
    num_cols = ['ShoppingMall','FoodCourt','RoomService','Spa','VRDeck','Age']
    cat_cols = ['CryoSleep','VIP','HomePlanet','Destination', ]

    num_imp = SimpleImputer(strategy='mean')
    cat_imp = SimpleImputer(strategy='most_frequent')
    
    df[num_cols] = pd.DataFrame(num_imp.fit_transform(df[num_cols]),columns=num_cols)
    df[cat_cols] = pd.DataFrame(cat_imp.fit_transform(df[cat_cols]),columns=cat_cols)
    return df
```

```python
handle_null_values(train)
handle_null_values(test)
```

## Drop values that don't matter

```python
def drop_values(df):
    df.drop('PassengerId', axis=1, inplace=True)
    df.drop('Name', axis=1, inplace=True)
    df.drop('Cabin', axis=1, inplace=True)
    df.drop('Destination', axis=1, inplace=True)
    return df

drop_values(train)
drop_values(test)
```

## Model Train

```python
from sklearn.model_selection import train_test_split, cross_val_score
categorical_columns = ['HomePlanet', 'CryoSleep', 'VIP']


X_train = train.drop('Transported', axis=1)
y_train = train['Transported']

X_test = test
X_train[categorical_columns] = X_train[categorical_columns].astype('category')
X_test[categorical_columns] = X_test[categorical_columns].astype('category')
```

```python
# Initialize the CatBoost classifier
model = CatBoostClassifier()
param = {
    'iterations': [100, 200, 300],
    'learning_rate': [0.01, 0.05, 0.1],
    'depth': [3, 5, 7],
    }
         
random = RandomizedSearchCV(estimator = model,param_distributions = param,verbose=3,cv=5,n_iter=5,scoring='neg_mean_squared_error', random_state=42)

random.fit(X_train, y_train, cat_features=categorical_columns)
```

```python
best_model = random.best_estimator_
best_params = random.best_params_

y_pred = best_model.predict(X_test)
```

```python
submission_df = pd.DataFrame({'PassengerId': passe, 'Transported': y_pred})
submission_df.to_csv('submission.csv', index=False)
```
